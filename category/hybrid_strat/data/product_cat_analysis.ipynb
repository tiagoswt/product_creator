```python
import pandas as pd
```

```python
# Load a CSV file into a DataFrame
df = pd.read_csv('product_info.csv', delimiter=";")

# Display the first few rows
df.head()
```

```python
df.info()
```

```python
print(df['encodeNameEN'].value_counts().to_string())
```

```python
# Define the unwanted categories
unwanted_categories = [
    'christmas/gift-sets',
    'gifting/perfumes',
    'christmas/gifts-premium',
    'gifting/premium',
    'gifting/secret-santa',
    'brands?d=3',
    'gifting',
    'gifting/for-her'
]

# Remove rows where 'encodeNameEN' matches any in the list
df_cleaned = df[~df['encodeNameEN'].isin(unwanted_categories)]
```

```python
df_cleaned
```

```python
print(df_cleaned['encodeNameEN'].value_counts().to_string())
```

```python
# Load a CSV file into a DataFrame
df_cat = pd.read_csv('product_description.csv', delimiter=";")

# Display the first few rows
df_cat.head()
```

```python
# Clean column names just to be sure
df_cleaned.columns = df_cleaned.columns.str.strip()
df_cat.columns = df_cat.columns.str.strip()

# Get unique categories from both DataFrames
cleaned_categories = set(df_cleaned['encodeNameEN'].dropna().unique())
cat_categories = set(df_cat['encodeNameEN'].dropna().unique())

# Find categories in df_cleaned but not in df_cat
missing_in_cat = cleaned_categories - cat_categories

# Filter out non-strings before sorting
missing_in_cat = [cat for cat in missing_in_cat if isinstance(cat, str)]

# Display result
print("Categories in df_cleaned but NOT in df_cat:")
for category in sorted(missing_in_cat):
    print(category)

print(f"\nTotal unmatched categories: {len(missing_in_cat)}")
```

```python
df_cleaned = df_cleaned.copy()

df_cleaned['encodeNameEN'] = df_cleaned['encodeNameEN'].replace(
    'body/moisturisers/dryness-dehydration', 
    'body/dehydration/moisturizers'
)
```

```python
print(df_cleaned['encodeNameEN'].value_counts().to_string())
```
